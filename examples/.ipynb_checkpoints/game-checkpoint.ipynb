# Instructions for the adjudicator package

A hands on user guide for using the adjudicator module to run a text based game of Diplomacy.

## Imports

Only run this cell once.

```python
import glob, os
os.chdir('../')  # Set working directory to main folder.

import adjudicator.game as gm
```

## Available variants

Before we start the game, let us list the available variants and maps.

At the time of this writing, there are three available variants and there are two available maps. The availble variants are:

1. Classic: The standard seven player version of the game.

2. ClassicFvA: The 1v1 version played between France and Austria on the standard map.

3. RPS: A 1v1 version on a map with three regions which is equivalent to Rock-Paper-Scissors.

```python
print(glob.glob('variants/*.json'))
print(glob.glob('maps/*.json'))
```

## Loading a game.

You start a game by initiating an instance of the gm.Game class, with the name of the variant as the only parameter. The name of the variant is the same as the name of the .json file. We will start with a game of RPS, as it is played on the smallest map.

Use the print function to get an overview of the current state of the game.

```python
game = gm.Game('RPS')
print(game)
```

As you can see, the game is in a _Pregame_ phase. In particular, units and ownership of home supply centers has not been set up. The variant and the map has been loaded, and their _static_ information may be accessed with the 'info' method of the gm.Game class.

The RPS variant is played in the RPS map. There are three provinces, all of which are supply centers. As you might tell, the variant plays out in West Africa. There are standard abbreviations for the three provinces. 

```python
game.info('variant')
game.info('map')
game.info('provinces')
game.info('supply centers')
game.info('abbreviations')
```

## Starting a game
You start the game with the start() method. This will set up home centers and units in starting positions. This _dynamic_ information may also be accessed via the display function. 

```python
game.start()
game.info('season')
game.info('units')
```

Here is the full list of options for the display function:

```python
game.info('options')
```

## Entering orders
Every unit is assigned a default Hold order. This might be a good time to point out some fundamental properties of the interface. At the beginning of each face, a list of default orders are generated. Each new order entered will replace one of the old orders.

```python
game.info('orders')
```

The orders are marked as unresolved, as the phase has not yet been adjudicated.

You enter orders as strings using the order() method of the gm.Game class. You can also enter a list of orders (i.e., a list of strings.)

```python
game.order('Army Conakry move to Monrovia.')
game.order('Army Monrovia move to Freetown.')
game.info('orders')
```

You can also enter order using the standard abbreviations.

```python
game.order(['A Con H.', 'A Mon - Con.'])
game.info('orders')
```

A lot can be said about the order input interface. The general principle is as follows: the method will do its best to interpret the string as a valid order, and to identify which order is to be replaced. It acts on a 'first seen' basis, meaning that excess information will be ignored. As long as you enter orders in any of the standard formats, there should not be a problem, but please check (using `game.info('orders')`) before adjudicating. 

N.B., some invalid orders will be accepted by the interface, but failed during the adjudication. Other invalid orders will be refused by the interface, implying that the previously stored order will be used during the adjudication. House rules regarding which orders are accepted might not agree with the principles of the adjudicator.

## Adjudicating
Once all orders are in, you can adjudicate the game using the `adjudicate()` method.

```python
game.adjudicate()
print(game)
```

As you can see, we have now moved in to the Fall season of 1990.

Displaying orders, at this point, will display the list of default order for the Fall turn. If you wish to see the outcome of the Spring adjudication, then you have to look inte the order archive.

```python
game.info('archive', 0)
```

The order archive already has two entries at this point. The second entry is for the 'Retreat' phase of Spring season of 1990. Since no units was forced to reateat (indeed, retreats are impossible in the RPS variant), this entry is an empty list.

Notice that since there are no retreats, the game moves on automatically to the Fall season. The same holds for the build phase; if there are no adjustments, then the game will move on automatically to the next Spring season.

Failed orders are marked with a '(fails)' tag, successful orders are not. In difference to some other interfaces, orders belonging to units which are forced to retreat are not marked (as being forced to retreat is a property of the unit, and not the order).

## Rolling back
If something went wrong (probably because an order was entered incorrectly), then you can roll back to the previous turn using the rollback() method.

You cannot roll back to the 'Pregame' phase.

```python
game.rollback()
print(game)
```

Rolling back will not overwrite the orders you entered for the last phase.

```python
game.info('orders')
```

## Graphics
Some variant includes the option to display a graphical map. At the time of this writing, only positions are displayed on the map. _This might change in the future._ You display the map by calling either `info('display')` or `show()`.

```python
game.show()
```

```python
game.order(['A Con - Fre.', 'A Mon - Con.'])
game.adjudicate()
game.info('archive', 0)
game.show()
```

```python
game.order(['A Con - Mon.', 'A Fre - Mon.'])
game.adjudicate()
game.info('archive', 2)
game.show()
```

## Support, Convoys, Retreats, Builds, Disbands 
In order to examplify the interface for the remaining orders, we will play a game of classic Diplomacy. A list of abbreviations is provided, as there are conflicting standards.

```python
game = gm.Game('Classic')
game.start()
game.info('abbreviations')
```

We will enter order power by power. Notice the French support into Burgundy. Notice that unordered units are recorded with the default hold order.

There is a minimalistic graphics package included also for the classic map. 

```python
game.order(['A Bud - Gal'])
game.order(['F Lon - NTH', 'A Lvp - Yor'])
game.order(['A Par - Bur', 'A Mar S Par - Bur'])
game.order(['A Ber - Sil', 'A Mun - Bur'])
game.order(['A War - Pru'])
game.order(['A Ven - Tyr'])
game.adjudicate()
game.show()
```

If an army is to move via convoy, then this must be specified. The standard abbreviation 'C' for convoy is accepted also in this setting.

Notice that the Italian Army in Tyrolia is supported into Munich by the French Army in Burgundy. The result is that Munich is forced to retreat. _The default order for a retreating unit is to disband._

```python
game.order(['A Yor - Nwy via C', 'F NTH C Yor - Nwy'])
game.order(['A Sil - War'])
game.order(['A Tyr - Mun'])
game.order(['A Bur S A Tyr - Mun'])
game.adjudicate()
game.show()
```

We enter a retreat order for Munisch, and move on to the build phase.

```python
game.order('A Mun - Boh')
game.adjudicate()
game.show()
```
